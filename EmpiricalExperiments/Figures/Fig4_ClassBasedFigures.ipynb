```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot
```

# Plots for Class-Based Homophily
Each plot is customized to the relevant dataset

## bills-house

```python
## Load Data
groups = pd.read_csv('../Results/bootstrapping_bills-house.csv')

# edit
groups = groups[groups['group_code']!=2]
#         print(groups)

# Set axes 
hb_vals = groups.loc[:,['h/b_type-1','h/b_type-2','h/b_type-3','sc_h/b_type-1','sc_h/b_type-2','sc_h/b_type-3']]
ymin = max(hb_vals.to_numpy().min() - 0.1, -1)
ymax = min(hb_vals.to_numpy().max() + 0.1, 100) 
```

```python
aff_base_m = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'h/b_type-1':'mean','h/b_type-2':'mean', 'h/b_type-3':'mean'}) #.set_index('group_code').transpose()

aff_base_s = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'h/b_type-1':'std','h/b_type-2':'std', 'h/b_type-3':'std'})

aff_base_sc_m = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'sc_h/b_type-1':'mean','sc_h/b_type-2':'mean', 'sc_h/b_type-3':'mean'})

aff_base_sc_s = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'sc_h/b_type-1':'std','sc_h/b_type-2':'std', 'sc_h/b_type-3':'std'})
```

```python
hyp_mean = pd.melt(aff_base_m, id_vars=['group_code'], value_vars=['h/b_type-1', 'h/b_type-2', 'h/b_type-3']).rename(columns={'value':'mean'})
hyp_std = pd.melt(aff_base_s, id_vars=['group_code'], value_vars=['h/b_type-1', 'h/b_type-2', 'h/b_type-3']).rename(columns={'value':'std'})
hyp = hyp_mean.merge(hyp_std, on=['group_code','variable'])


sim_mean = pd.melt(aff_base_sc_m, id_vars=['group_code'], value_vars=['sc_h/b_type-1', 'sc_h/b_type-2', 'sc_h/b_type-3']).rename(columns={'value':'mean'})
sim_std = pd.melt(aff_base_sc_s, id_vars=['group_code'], value_vars=['sc_h/b_type-1', 'sc_h/b_type-2', 'sc_h/b_type-3']).rename(columns={'value':'std'})
sim = sim_mean.merge(sim_std, on=['group_code','variable'])

```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10, 4))

# 0 - Democrat, 1- Republican
labels = {0:'Democrat', 1:'Republican'}
colors = {0:'blue', 1:'red'}
for key, group in hyp.groupby('group_code'):
    group.plot('variable', 'mean', yerr='std', label=labels[key], ax=axes[0], color = colors[key])
    axes[0].fill_between(group['variable'], group['mean']-group['std'], 
                         group['mean']+group['std'], alpha=0.1, facecolor=colors[key])

for key, group in sim.groupby('group_code'):
    group.plot('variable', 'mean', yerr='std', label=labels[key], ax=axes[1], color = colors[key])
    axes[1].fill_between(group['variable'], group['mean']-group['std'], 
                         group['mean']+group['std'], alpha=0.1, facecolor=colors[key])

axes[0].hlines(1, 0, 2, color='black', linestyle='--')
axes[0].set_ylim([ymin, ymax])
axes[0].set_xlabel('Affinity type t')#, fontweight='bold')
axes[0].set_title('Hypergraph Homophily')#, fontweight='bold')
axes[0].set_yscale('log')
axes[0].legend(title = 'Political Party')
axes[0].set_xticks([0, 1, 2])
axes[0].set_xticklabels(['type-1', 'type-2', 'type-3'])
[x.set_linewidth(1.5) for x in axes[0].spines.values()]
# axes[0].set_ylabel('(a)', fontsize = 18, fontweight='bold', rotation=0)


axes[1].hlines(1, 0, 2, color='black', linestyle='--')
axes[1].set_ylim([ymin, ymax])
axes[1].set_xlabel('Affinity type t')#, fontweight='bold')
axes[1].set_title('2-Simplicial Homophily')#, fontweight='bold')
axes[1].set_yscale('log')
axes[1].legend(title = 'Political Party')
axes[1].set_xticks([0, 1, 2])
axes[1].set_xticklabels(['type-1', 'type-2', 'type-3'])
[x.set_linewidth(1.5) for x in axes[1].spines.values()]

fig.savefig('bills-house.png', dpi=300)
```

## cont-village
(*) Because this dataset is small, we find that the bootstrapping procedure sometimes leads to fewer than 3 nodes of a class being sampled, making type-3 class-based homophily impossible to compute

```python
## Load Data
groups = pd.read_csv('../Results/bootstrapping_cont-village.csv')
groups = groups.loc[:,['group_code','h/b_type-1','h/b_type-2','h/b_type-3','sc_h/b_type-1','sc_h/b_type-2','sc_h/b_type-3']]
# groups.replace([np.inf, -np.inf], np.nan, inplace=True)
# groups = groups.dropna()

# edit
#         print(groups)

# Set axes 
hb_vals = groups.loc[:,['h/b_type-1','h/b_type-2','h/b_type-3','sc_h/b_type-1','sc_h/b_type-2','sc_h/b_type-3']]
ymin = max(hb_vals.to_numpy().min() - 0.1, -1)
ymax = min(hb_vals.to_numpy().max() + 0.1, 100) 
```

```python
groups
```

```python
aff_base_m = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'h/b_type-1':'mean','h/b_type-2':'mean', 'h/b_type-3':'mean'}, dropna=True) #.set_index('group_code').transpose()

aff_base_s = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'h/b_type-1':'std','h/b_type-2':'std', 'h/b_type-3':'std'},  dropna=True)

aff_base_sc_m = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'sc_h/b_type-1':'mean','sc_h/b_type-2':'mean', 'sc_h/b_type-3':'mean'},  dropna=True)

aff_base_sc_s = groups.groupby(by=['group_code'], as_index=False).agg(
                      {'sc_h/b_type-1':'std','sc_h/b_type-2':'std', 'sc_h/b_type-3':'std'},  dropna=True)
```

```python
hyp_mean = pd.melt(aff_base_m, id_vars=['group_code'], value_vars=['h/b_type-1', 'h/b_type-2', 'h/b_type-3']).rename(columns={'value':'mean'})
hyp_std = pd.melt(aff_base_s, id_vars=['group_code'], value_vars=['h/b_type-1', 'h/b_type-2', 'h/b_type-3']).rename(columns={'value':'std'})
hyp = hyp_mean.merge(hyp_std, on=['group_code','variable'])


sim_mean = pd.melt(aff_base_sc_m, id_vars=['group_code'], value_vars=['sc_h/b_type-1', 'sc_h/b_type-2', 'sc_h/b_type-3']).rename(columns={'value':'mean'})
sim_std = pd.melt(aff_base_sc_s, id_vars=['group_code'], value_vars=['sc_h/b_type-1', 'sc_h/b_type-2', 'sc_h/b_type-3']).rename(columns={'value':'std'})
sim = sim_mean.merge(sim_std, on=['group_code','variable'])

```

```python
hyp
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize = (10, 4))

# 0 - Democrat, 1- Republican
labels = {0 : 'E', 1 : 'F', 2 : 'L', 3 : 'B', 4 : 'H'}
colors = {0:'xkcd:teal', 1:'xkcd:olive', 2:'xkcd:green', 3:'xkcd:magenta', 4:'xkcd:plum'}
for key, group in hyp.groupby('group_code'):
    group.plot('variable', 'mean', yerr='std', label=labels[key], ax=axes[0], color = colors[key])
    axes[0].fill_between(group['variable'], group['mean']-group['std'], 
                         group['mean']+group['std'], alpha=0.2, facecolor=colors[key])

for key, group in sim.groupby('group_code'):
    group.plot('variable', 'mean', yerr='std', label=labels[key], ax=axes[1], color = colors[key])
    axes[1].fill_between(group['variable'], group['mean']-group['std'], 
                         group['mean']+group['std'], alpha=0.2, facecolor=colors[key])

axes[0].hlines(1, 0, 2, color='black', linestyle='--')
axes[0].set_ylim([0.1, 8.0])
axes[0].set_xlabel('Affinity type t', fontweight='bold')
axes[0].set_title('Hypergraph Homophily', fontweight='bold')
axes[0].set_yscale('log')
axes[0].legend(title = 'Village')
axes[0].set_xticks([0, 1, 2])
axes[0].set_xticklabels(['type-1', 'type-2', 'type-3'])
[x.set_linewidth(1.5) for x in axes[0].spines.values()]
# axes[0].set_ylabel('house-bills', fontweight='bold', rotation=0)


axes[1].hlines(1, 0, 2, color='black', linestyle='--')
axes[1].set_ylim([0.1, 8.0])
axes[1].set_xlabel('Affinity type t', fontweight='bold')
axes[1].set_title('2-Simplicial Homophily', fontweight='bold')
axes[1].set_yscale('log')
axes[1].legend(title = 'Village')
axes[1].set_xticks([0, 1, 2])
axes[1].set_xticklabels(['type-1', 'type-2', 'type-3'])
[x.set_linewidth(1.5) for x in axes[1].spines.values()]

fig.savefig('cont-village.png', dpi=300)
```
